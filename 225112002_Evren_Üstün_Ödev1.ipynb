## Soru 1:

9 farklı renkte balon şekildeki gibi asılmıştır. Bu balona atış yapacak olan atıcı ipteki balonların en alttakini patlatmadan üstteki balona atış yapamamaktadır.

image.png

Buna göre, her atışında bir balon patlatmak şartıyla atıcı balonların tamamını kaç farklı sıralamayla vurulabilir?

## Çözüm:

Birinci sütunda bulunan balonlara S1, <br>
İkinci sütunda bulunan balonlara S2, <br>
Üçüncü sütunda bulunan balonlara S3 ile isimlendirelim. <br>

Atışçı S1, S2, S1, S2, S3, S3, S2, S3, S1 şeklinde atışlarını gerçekleştirebilir. <br>

9!/(3! * 3! * 3!) = 1680 <br>

(3! * 3! * 3!) yapmamızın sebebi 3 tane S1'de , 3 tane S2'de ve 3 tane S3'te aynı sıralama olacağı için aynı sıralamaları ortadan kaldırmak adına tekrarlı permütasyondan faydalandık. <br>

Atışçı 1680 farklı sıralama ile balonları vurabilir.


## Pyhthon Kod Çözümü:

```python
def faktoriyelHesapla(i):    
  if i==1:       
    return 1
  else: 
    return i * faktoriyelHesapla (i-1)   

sonuc = faktoriyelHesapla(9) / (faktoriyelHesapla(3) * faktoriyelHesapla(3) * faktoriyelHesapla(3)) 
 

print('Atışçı {} farklı sıralama ile balonları vurabilir.'.format(int(sonuc)))
```

## Soru 2:

Ankara'dan İstanbul'a giden 8 kişilik ekip Düzce'de rafting yapmak istemektedir. <br>

Eğitmen haricinde 4 kişi taşıyabilen iki bota Binmeyi planlamaktadır. <br>

Buna göre, 8 kişi dört kişilik iki farklı bota kaç farklı şekilde binebilir?

## Çözüm:
Birinci bota 1 eğitmen ve 8 kişiden 4 kişi binecektir.
kalan 4 kişinin dördü de ikinci bota binecektir. Bir de ikinci bottakiler 1. bota veya 1. bottaki kişiler ikinci botta olabilir. Bundan dolayı cevap: <br>

C(8,4) * C(4,4) * 2! = 140



## Pyhton Kod Çözümü:

```python
def faktoriyelHesapla(i):    
  if i==1:       
    return 1
  else: 
    return i * faktoriyelHesapla (i-1)   

def kombinasyonHesapla(ilksayi, ikincisayi):
     
    if (ilksayi<ikincisayi):
        print("1. Sayının 2. Sayıdan büyük yada eşit olması gerekli. Verdiğiniz sayıların çözümü yoktur!")
    
    elif(ilksayi==ikincisayi):
        return 1
     
    elif(ilksayi>ikincisayi):
         
        donen1=1
        donen1= faktoriyelHesapla(ilksayi)
         
        donen2=1
        donen2= faktoriyelHesapla(ikincisayi) 
         
        donen3=1
        donen3= faktoriyelHesapla(ilksayi-ikincisayi)
         
        sonuc=donen1/(donen2*donen3)
        return sonuc


sonuc =   kombinasyonHesapla(8, 4) * kombinasyonHesapla(4, 4) * faktoriyelHesapla(2)

print('8 kişi dört kişilik iki farklı bota {} farklı şekilde binebilir.'.format(int(sonuc)))       
```
